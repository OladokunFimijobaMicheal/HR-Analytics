```python
# importing the libraries
import pandas as pd
```

```python
# loading the data
df = pd.read_csv(r"C:\Users\Apex PC\Downloads\HR Employee\employee_data.csv")
df
```

```python
# displaying all the columns
pd.set_option('display.max.columns', 28)
```

```python
# renaming the columns
df.rename(columns={
    'EmpID' : 'Employee ID',
    'StartDate' : 'Start Date',
    'ExitDate' : 'Exit Date',
    'ADEmail' : 'Mail',
    'EmployeeStatus' : 'Employee Status',
    'JobFunctionDescription' : 'Job',
    'DOB' : 'Date of Birth',
    'GenderCode' : 'Gender',
    'LocationCode' : 'Zipcode',
    'MaritalDesc' : 'Marital Status', 
    'EmployeeType' : 'Job Type',
    'PayZone' : 'Zone',
    'EmployeeClassificationType' : 'Contract Type',
    'TerminationType' : 'Termination Type',
    'TerminationDescription' : 'Termination Description',
    'DepartmentType' : 'Department',
    'FirstName' : 'First Name',
    'LastName' : 'Last Name',
    'BusinessUnit' : 'Business Unit'
    }, inplace=True)
```

```python
# joining the columns
df ['Full Name'] = df[['First Name', 'Last Name']].apply(lambda x: ' '.join(x), axis=1)
df
```

```python
# dropping irrelevant columns
df.drop(columns=['Termination Description', 'First Name', 'Last Name', 'Exit Date'], inplace=True)

df
```

```python
# moving full name to index 1, axis 1 i.e beside Employee OD
df = df.loc[:,[
'Employee ID', 'Full Name', 'Start Date' ,	'Title', 'Supervisor',
'Mail',	'Business Unit', 'Employee Status',	'Job Type',	'Zone',	'Contract Type',
'Termination Type',	'Department', 'Division', 'Date of Birth', 'State',	'Job',
'Gender', 'Zipcode','RaceDesc',	'Marital Status','Performance Score', 'Current Employee Rating'
]]
```

```python
# Renaming the values in state column
df['State'].replace('AL', 'Alabama', inplace=True)
df['State'].replace('AZ', 'Arizona', inplace=True)
df['State'].replace('CA', 'California', inplace=True)
df['State'].replace('CO', 'Colorado', inplace=True)
df['State'].replace('CT', 'Connecticut', inplace=True)
df['State'].replace('FL', 'Florida', inplace=True)
df['State'].replace('GA', 'Georgia', inplace=True)
df['State'].replace('ID', 'Idaho', inplace=True)
df['State'].replace('IN', 'Indiana', inplace=True)
df['State'].replace('KY', 'Kentucky', inplace=True)
df['State'].replace('MA', 'Massachussets', inplace=True)
df['State'].replace('ME', 'Maine', inplace=True)
df['State'].replace('MT', 'Montana', inplace=True)
df['State'].replace('NC', 'North Carolina', inplace=True)
df['State'].replace('ND', 'North Dakota', inplace=True)
df['State'].replace('NH', 'New Hamphshire', inplace=True)
df['State'].replace('NV', 'Nevada', inplace=True)
df['State'].replace('NY', 'New York', inplace=True)
df['State'].replace('OH', 'Ohio', inplace=True)
df['State'].replace('OR', 'Oregon', inplace=True)
df['State'].replace('PA', 'Pennsylvania', inplace=True)
df['State'].replace('RI', 'Rhode Island', inplace=True)
df['State'].replace('TN', 'Tennesse', inplace=True)
df['State'].replace('TX', 'Texas', inplace=True)
df['State'].replace('UT', 'Utah', inplace=True)
df['State'].replace('VA', 'Virginia', inplace=True)
df['State'].replace('VT', 'Vermont', inplace=True)
df['State'].replace('WA', 'Washington', inplace=True)
```

```python
df['Termination Type'].replace('Unk', 'Unknown', inplace=True)
df['Department'].replace('IT/IS', 'IT', inplace=True)
df['Division'].replace('Technology / It', 'Technology', inplace=True)
df['Division'].replace('Catv', 'Cable Television', inplace=True)
df['Division'].replace('Corp Operations', 'Cooperate ', inplace=True)
df['Division'].replace('General - Sga', 'General - Selling General and Administrative', inplace=True)
df['Division'].replace('General - Con', ' General - Conference', inplace=True)
df['Division'].replace('General - Eng', 'General - Engineering', inplace=True)
df['Division'].replace('Isp', 'Internet Service Provider', inplace=True)
df['Job'].replace('Cfo', 'Chief Financial Officer', inplace=True)
df['Job'].replace('Cio', 'Chief Informaion Officer', inplace=True)
df['Job'].replace('Cpo', 'Chief Product Officer', inplace=True)
df['Job'].replace('Evp', 'Executive Vice President', inplace=True)
df['Job'].replace('Svp', 'Senior Vice President', inplace=True)
df['Job'].replace('Vp', 'Vice President', inplace=True)
df['Job'].replace('Ceo', 'Chief Financial Officer', inplace=True)
```

```python
# dropping duplicates
df.drop_duplicates()
```

```python
# dropping null values
df.dropna()
```
