```python
# importing the libraries
import pandas as pd
```

```python
# loading the data
df = pd.read_csv(r"C:\Users\Apex PC\Downloads\HR Employee\recruitment_data.csv")
df
```

```python
# replacing values
df['Phone Number'] = df['Phone Number'].str.replace('x', '-', regex=True)
df['Phone Number'] = df['Phone Number'].str.replace('+', '', regex=True)
df['Phone Number'] = df['Phone Number'].str.lstrip('00')

df
```

```python
# dropping all the brackets
df['Phone Number'] = df['Phone Number'].str.strip('()')
df['Phone Number'] = df['Phone Number'].str.replace(')', '', regex=True)
df['Phone Number'] = df['Phone Number'].str.replace('(', '', regex=True)
df['Phone Number']

```

```python
#removes all characters that are not numbers
df['Phone Number'] = df['Phone Number'].str.replace('[^a-zA-Z0-9]', '', regex=True) 
df['Phone_Number']= df['Phone Number'].apply(lambda x:str(x))
```

```python
# adds - at the specified indexes
df['Phone Number'] = df['Phone Number'].apply(lambda x:x[:0] + '+' + x[3:6 ]+ '-' + x[6:10 ]  + x[10:13 ]  + x[13:17] )
df['Phone Number']
```
